# 04. 元素記号
“Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can.”という文を単語に分解し，1, 5, 6, 7, 8, 9, 15, 16, 19番目の単語は先頭の1文字，それ以外の単語は先頭の2文字を取り出し，取り出した文字列から単語の位置（先頭から何番目の単語か）への連想配列（辞書型もしくはマップ型）を作成せよ．

```python
str = 'Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can.'
str_s = str.split()
str_s = [s.strip(',.') for s in str_s]
# print(str_s)
dic = {}
for i in range(len(str_s)):
  if i+1 in [1, 5, 6, 7, 8, 9, 15, 16, 19]:
    dic[i+1] = str_s[i][0]
  else:
    dic[i+1] = str_s[i][:2]

print(dic)
```

# 05. n-gram
与えられたシーケンス（文字列やリストなど）からn-gramを作る関数を作成せよ．この関数を用い，”I am an NLPer”という文から単語bi-gram，文字bi-gramを得よ．


n-gram とは n 個のデータの連接である．  
1-gram, 2-gram, 3-gram, ..., n-gram となる．
一般に 1-gram, 2-gram, 3-gram をそれぞれ uni-gram, bi-gram, tri-gram と書く．  
それぞれ「ユニグラム」，「バイグラム」，「トライグラム」と読む．  
単語 n-gram は単語を単位とした連接，文字 n-gram は文字を単位とした連接である．  
"I saw a girl with a telescope." という文があったとしたら，  
単語 uni-gram は "I", "saw", "a",...,"telescope."，  
文字 uni-gram は "I", " ", "s", "a", "w",...,"."，  
単語 bi-gram は "I saw", "saw a", "a girl",..., "a telescope."，  
文字 bi-gram は "I ", " s", "sa", "aw", "w ", " a",...,"e."  
となる．

```python
def n_gram(str, n):
  wo_gram = []
  le_gram = []
  wo = ''
  le = ''
  str_s = str.split()
  # str_s = [s.strip(',.') for s in str_s] 例文的にしないっぽい

  # 単語
  for i in range(len(str_s)):
    for j in range(n):
      if i + j >= len(str_s):
        wo = ''
        break
      if wo != '':
        wo += ' '
      wo +=str_s[i+j]
    if wo != '':
      wo_gram.append(wo)
    wo = ''

  # 文字
  for i in range(len(str)):
    for j in range(n):
      if i + j >= len(str):
        le = ''
        break
      le += str[i+j]
    if le != '':
      le_gram.append(le)
    le = ''

  # print(str)
  # print(str_s)
  return wo_gram, le_gram

if __name__ == '__main__':
  str = 'I am an NLPer'
  n = 2
  result = n_gram(str, n)
  print('文字列', str, 'の')
  print('単語', n, '-gramは')
  print(result[0])
  print('文字', n, '-gramは')
  print(result[1])
  print('となる')
```

# 06. 集合
“paraparaparadise”と”paragraph”に含まれる文字bi-gramの集合を，それぞれ, XとYとして求め，XとYの和集合，積集合，差集合を求めよ．さらに，’se’というbi-gramがXおよびYに含まれるかどうかを調べよ．


```python
str_x = 'paraparaparadise'
str_y = 'paragraph'
X = set(n_gram(str_x, 2)[1])
Y = set(n_gram(str_y, 2)[1])

Uni = X | Y
Int = X & Y
Dif = X - Y

print('Xは')
print(X)
print('Yは')
print(Y)
print()
print('XとYの')
print('和集合は')
print(Uni)
print('積集合は')
print(Int)
print('差集合は')
print(Dif)
print('となる')
print()
print('seというbi-gramは')

if 'se' in X:
  print('Xに含まれ')
else:
  print('Xに含まれず')

if 'se' in Y:
  print('Yに含まれる')
else:
  print('Yに含まれない')
```

# 08. 暗号文
与えられた文字列の各文字を，以下の仕様で変換する関数cipherを実装せよ．

英小文字ならば(219 - 文字コード)の文字に置換
その他の文字はそのまま出力
この関数を用い，英語のメッセージを暗号化・復号化せよ．

```python
def cipher(s):
  a = 'a'
  z = 'z'
  ord_a = ord(a)
  ord_z = ord(z)
  # print(ord_a)
  # print(ord_z)
  s_r = ''
  for ss in s:
    ord_ss = ord(ss)
    if ord_a <= ord_ss <= ord_z:
      ord_ss_2 = 219 - ord_ss
      ss = chr(ord_ss_2)
    s_r += ss
  return s_r

if __name__ == '__main__':
  s = 'This is a pen.'
  print(s)
  print(cipher(s))
  print(cipher(cipher(s)))
```

# 09. Typoglycemia
スペースで区切られた単語列に対して，各単語の先頭と末尾の文字は残し，それ以外の文字の順序をランダムに並び替えるプログラムを作成せよ．ただし，長さが４以下の単語は並び替えないこととする．適当な英語の文（例えば”I couldn’t believe that I could actually understand what I was reading : the phenomenal power of the human mind .”）を与え，その実行結果を確認せよ．

```python
import random

s = 'I couldn’t believe that I could actually understand what I was reading : the phenomenal power of the human mind .'
print(s)
s_s = s.split()
# print(s_s)
s_s_r = []

for ss in s_s:
  if len(ss) <= 4:
    s_s_r.append(ss)
    continue
  ss_t = ss[0]
  ss_b = ss[len(ss)-1]
  ss_r = ss[1:len(ss)-1]
  ss_r = ''.join(random.sample(ss_r, len(ss_r)))
  ss = ss_t + ss_r + ss_b
  s_s_r.append(ss)

s_s_s = ' '.join(s_s_r)
print(s_s_s)
```
