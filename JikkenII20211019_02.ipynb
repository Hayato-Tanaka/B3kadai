# 10. 行数のカウント

# 11. タブをスペースに置換
タブ1文字につきスペース1文字に置換せよ．確認にはsedコマンド，trコマンド，もしくはexpandコマンドを用いよ．


```python
fa = open('./popular-names_2.txt', 'w')
fa.close()
with open('./popular-names.txt', 'r') as f:
  for data in f:
    with open('./popular-names_2.txt', 'a') as fa:
      print(data.strip().replace("\t"," "), file = fa)

!cat ./popular-names.txt | tr '\t' ' ' > ./popular-names_2_co.txt
```

# 12. 1列目をcol1.txtに，2列目をcol2.txtに保存
各行の1列目だけを抜き出したものをcol1.txtに，2列目だけを抜き出したものをcol2.txtとしてファイルに保存せよ．確認にはcutコマンドを用いよ．


```python
for i in  range(2):
  save = './col'
  save += str(i+1)
  save += '.txt'
  ff = open(save, 'w')
  ff.close()
  with open('./popular-names.txt', 'r') as f:
    for data in f:
      slist = data.split('\t')
      with open(save, 'a') as ff:
        print(slist[i], file = ff)

!cut -f 1 ./popular-names.txt > ./popular-names_1retsu.txt
!cut -f 2 ./popular-names.txt > ./popular-names_2retsu.txt
```

# 13. col1.txtとcol2.txtをマージ
12で作ったcol1.txtとcol2.txtを結合し，元のファイルの1列目と2列目をタブ区切りで並べたテキストファイルを作成せよ．確認にはpasteコマンドを用いよ．


```python
with open('./col1.txt', 'r') as f:
  col1 = f.readlines()

with open('./col2.txt', 'r') as f:
  col2 = f.readlines()

for i in range(len(col1)):
  col1[i] = col1[i].replace('\n', '')

for i in range(len(col2)):
  col2[i] = col2[i].replace('\n', '')

f = open('./col_merge.txt', 'w')
f.close()
for i in range(len(col1)):
  s_m = col1[i] + '\t' + col2[i]
  with open('./col_merge.txt', 'a') as f:
    print(s_m, file = f)

!paste -d \t ./col1.txt ./col2.txt > col_merge_com.txt
```

#16. ファイルをN分割する
自然数Nをコマンドライン引数などの手段で受け取り，入力のファイルを行単位でN分割せよ．同様の処理をsplitコマンドで実現せよ．


```python
N = 5

with open('./popular-names.txt', 'r') as f:
  l = f.readlines()

r = len(l) / 5
r_i = int(r)
for i in range(N):
  f_n = './popular_'
  f_n += str(i+1)
  f_n += '_'
  f_n += str(N)
  f_n += '.txt'
  f = open(f_n, 'w')
  f.close()
  with open(f_n, 'a') as f:
    print(''.join(l[i*r_i : (i+1)*r_i]), file = f)

!split -l 556 -d ./popular-names.txt ./popular_com.txt.
```

# 18. 各行を3コラム目の数値の降順にソート
各行を3コラム目の数値の逆順で整列せよ（注意: 各行の内容は変更せずに並び替えよ）．確認にはsortコマンドを用いよ（この問題はコマンドで実行した時の結果と合わなくてもよい）．


```python
from operator import itemgetter
fir = []
sec = []
thi = []
fou = []
with open('./popular-names.txt', 'r') as f:
  for data in f:
    slist = data.split('\t')
    fir.append(slist[0].strip())
    sec.append(slist[1].strip())
    thi.append(int(slist[2].strip()))
    fou.append(int(slist[3].strip()))
l = zip(fir, sec, thi, fou)
l_r = sorted(l, key = itemgetter(2), reverse = True)
fir, sec, thi, fou = zip(*l_r)
f = open('popular-names_sorted.txt', 'w')
f.close()
for i in range(len(fir)):
  s = fir[i] + '\t' + sec[i] + '\t' + str(thi[i]) + '\t' + str(fou[i])
  with open('popular-names_sorted.txt', 'a') as f:
    print(s, file = f)

!sort -k 3nr ./popular-names.txt >./popular-names_sorted_com.txt
```

# 19. 各行の1コラム目の文字列の出現頻度を求め，出現頻度の高い順に並べる
各行の1列目の文字列の出現頻度を求め，その高い順に並べて表示せよ．確認にはcut, uniq, sortコマンドを用いよ．

```python
import collections

with open('col1.txt', 'r') as f:
  l = f.readlines()

l_c = collections.Counter(l)
#print(l_c)
l_c_2 = l_c.most_common()
#print(l_c_2)

a = []
b = []
for i in range(len(l_c_2)):
  a.append(l_c_2[i][0])
for s in a:
  b.append(s.strip())
print(b)

!sort col1.txt | uniq -c | sort -nr > test.txt
```
